```python
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS

# Step 1: Set up the HuggingFace Embedding model (CPU)
embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2",
    model_kwargs={'device': 'cpu'},
    encode_kwargs={'normalize_embeddings': False}
)

# Step 2: Load your previously saved FAISS vector store
vectorstore = FAISS.load_local(
    "multi_format_faiss_index",
    embeddings,
    allow_dangerous_deserialization=True  # Required to load locally saved index
)

# Step 3: Confirm successful loading
print("âœ… FAISS index loaded successfully.")



```
